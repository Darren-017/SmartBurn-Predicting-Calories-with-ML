Model Training

```python

features = calories_data.drop(['User_ID', 'Calories'], axis=1)
target = calories_data['Calories'].values
X_train, X_val, Y_train, Y_val = train_test_split(features, target, test_size=0.1, random_state=22)
X_train.shape, X_val.shape
```

```python
# Normalizing the features for stable and fast training.
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)
```

```python
from sklearn.metrics import mean_absolute_error as mae
models = [LinearRegression(), XGBRegressor(),
          Lasso(), RandomForestRegressor(), Ridge()]

for i in range(5):
    models[i].fit(X_train, Y_train)

    print(f'{models[i]} : ')

    train_preds = models[i].predict(X_train)
    print('Training Error : ', mae(Y_train, train_preds))

    val_preds = models[i].predict(X_val)
    print('Validation Error : ', mae(Y_val, val_preds))
    print()
```

```python
for model in models:
    model.fit(X_train, Y_train)

    print(f'{model} : ')

    train_preds = model.predict(X_train)
    train_mae = mean_absolute_error(Y_train, train_preds)
    train_r2 = r2_score(Y_train, train_preds)
    print('Training Error (MAE): ', train_mae)
    print('Training Accuracy (R-squared): ', train_r2)

    val_preds = model.predict(X_val)
    val_mae = mean_absolute_error(Y_val, val_preds)
    val_r2 = r2_score(Y_val, val_preds)
    print('Validation Error (MAE): ', val_mae)
    print('Validation Accuracy (R-squared): ', val_r2)
    print()
```
